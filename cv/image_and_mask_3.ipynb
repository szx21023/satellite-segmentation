```python
class Mask:
    def __init__(self, image, clusters):
        self.image = image
        self.clusters = clusters
        
    def pixel_cluster(self, img, clusters):
        X = img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
        k_means = cluster.KMeans(n_clusters=clusters)
        k_means.fit(X)

        X_cluster = k_means.labels_
        X_cluster = X_cluster.reshape(img.shape[0], img.shape[1])
        return X_cluster

    def cluster_to_df(self):
        img = self.image
        clusters = self.clusters
        
        X_cluster = self.pixel_cluster(img, clusters)
        return pd.DataFrame(X_cluster)
```

```python
mask = Mask(img, 2)
```

```python
df = mask.cluster_to_df()
mask = df.values
mask = np.where(mask==1, 0, 1)
```
